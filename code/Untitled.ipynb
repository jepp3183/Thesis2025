```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import seaborn as sns
```

```python
X = make_blobs()
X[0].shape
plt.scatter(X[0][:,0], X[0][:,1])
```

```python
km = KMeans(n_clusters=3).fit_predict(X[0])
km
```

```python
plt.scatter(X[0][:,0], X[0][:,1], )
```

```python
sns.scatterplot(x=X[0][:,0], y=X[0][:,1], hue=km, palette=['red','green','blue'])
```
